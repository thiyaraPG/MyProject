```python
# extract file from zip file
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall() #extract file or dir
zip_ref.close() #close file
```

```python
#read data from file and handle null values
import pandas as pd
df = pd.read_csv('orders.csv', na_values=['Not Available', 'unknown']) #remove Not Available and unkonown class
df.head(20)
df['Ship Mode'].unique()  #distinct values available
```

```python
#rename column names..make them lower case and replace space with undercore
#df.rename(columns = {'Order Id' : 'order_id', 'City':'city'})
#df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_' )
df.head(5)
```

```python
#derive new columns discount, sale price and profit
#df['discount']=df['list_price']*df['discount_percent']*.01
df['sale_price'] = df['list_price'] - df['discount']
df['profit'] = df['sale_price'] - df['cost_price']
df
```

```python
#convert order data from object data type to datetime
df.dtypes
```

```python
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%M-%d")
df
```

```python
#drop cost price and discount percent columns
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
df
```
